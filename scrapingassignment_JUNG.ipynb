```python
labor_data = []

for url in urls_lc:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    print(f"Scraping {url}")
    chart = soup.find("div", class_ = "search-main-wrapper")
    case_files = chart.find_all("div", class_ = "wrapper-div")

    for files in case_files:
        company_name = files.find('a').get_text()
        extra_stuff = files.find("div", class_ = "left-div")
        case_number = files.find("a").get_text()
   
    
    labor_data.append({
        'Company Name': company_name,
        'Case Number' : case_number,
        
    })
```

```python
df = pd.DataFrame(labor_data)

df
```
